```python
#Importing the Important Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading the dataset with the help of the pandas library

data = pd.read_csv('dataset.csv')

#Showing the table
data
```

```python

# Set the size of the entire plot (width=12, height=7 ) for better visibility
plt.figure(figsize = (10 ,8))

#Plotting the graph between "Score" and "Social Support"
plt.plot(data["Score"], data["Social support"],    label="Social Support" ,) 
# Plotting the graph between "Score" and "Healthy life expectancy" with red dotted line and markers
plt.plot(data["Score"], data["Healthy life expectancy"], 'r.-'  , label="Healthy Life Expectancy")
# Plotting the graph between "Score" and "Generosity" with green dotted line and markers
plt.plot(data["Score"], data["Generosity"], 'g.-' ,  label="Generosity")

# Set label for the X-axis to indicate what is being measured along the horizontal axis
plt.xlabel("Score")
# Set label for the Y-axis to indicate the values of Social Support, Healthy Life Expectancy, and Generosity
plt.ylabel("Values")
# Set the title of the graph to describe the relationship being visualized
plt.title("Score vs Social Support , Health Life Expectancy & Generosity")
# Display a legend to distinguish between the lines representing different columns   
plt.legend()
#Render the Plot
plt.show()
```

```python
# Set the size of the entire plot (width=12, height=7) for better visibility
plt.figure(figsize=(12,7))

# Select all columns starting from the 4th one (index 3 onward) to be plotted
# This assumes the first three columns are not relevant for this plot           
columns_to_plot = data.columns[3:]
           
# Iterate through each selected column to plot its values against the "Score" column
# This loop is necessary to dynamically plot multiple lines on the same graph
# Instead of hardcoding each line, the loop makes it scalable and cleaner   
           
for column in columns_to_plot:# Plot the current column's values vs. Score with a dot marker and label it
    plt.plot(data["Score"], data[column], marker='.', label=column)
         
# Set label for the X-axis           
plt.xlabel("Score")
# Set label for the Y-axis           
plt.ylabel("Values")
# Set the title of the graph with bold font and larger size for emphasis           
plt.title("Comparison of Different Factors Over Years",fontdict = {'fontweight':'bold','fontsize':22})
# Display a legend to distinguish between the lines representing different columns           
plt.legend()
# Render the plot           
plt.show()
```

```python
# Set the size of the entire plot (width=8, height=5) for better visibility

plt.figure(figsize=(8,5))

# X-Axis: GDP per Capita
# I have specified bins=15, which tells matplotlib to divide the range of GDP per capita values into 15 equal-width intervals (or "bins" / "rectangular boxes").
# The x-axis shows the edges of these bins.
# matplotlib calculates the minimum and maximum values in the data["GDP per capita"] column, then splits that range into 15 segments.
# Each tick on the x-axis represents the start of a bin.

# Y-Axis: Frequency
# The y-axis shows the count of data points (how many countries) that fall within each bin.
plt.hist(data["GDP per capita"], bins=15, color='#aabcdd', edgecolor='black')

#Set the label for the X-axis
plt.xlabel("GDP per Capita")
#Set the label for thr Y-axis
plt.ylabel("Frequency")
#Setting the title of the graph plotting
plt.title("Histogram of GDP per Capita")



#Render the Plot
plt.show()
```

```python
# Extract the GDP per capita value for Norway
norway_gdp = data[data['Country or region'] == 'Norway']['GDP per capita']

# Extract the GDP per capita value for Netherlands
netherlands_gdp = data[data['Country or region'] == 'Netherlands']['GDP per capita']

# Define the country names to be used as labels on the x-axis
labels = ['Norway', 'Netherlands']

# Create a box plot to compare the GDP per capita between Norway and Netherlands
# Each element in the list corresponds to the data for one country
plt.boxplot([norway_gdp, netherlands_gdp], labels=labels)

# Set the title of the box plot
plt.title('GDP per Capita Comparison: Norway vs Netherlands')

# Label the y-axis to indicate the values represent GDP per capita
plt.ylabel('GDP per Capita')

# Render and display the box plot
plt.show()
```

```python
# Extracting the data row for the country 'Norway'
# The result is a single row (as a Series), representing all factor values for Norway

Norway = data[data['Country or region'] == 'Norway'].iloc[0]
# Creating a dictionary to store selected happiness factors and their corresponding values for Norway
# These factors represent the key contributors to the overall happiness score

factors = {
    'GDP': Norway['GDP per capita'],
    'Social Support': Norway['Social support'],
    'Life Expectancy': Norway['Healthy life expectancy'],
    'Freedom': Norway['Freedom to make life choices'],
    'Generosity': Norway['Generosity'],
    'Corruption': Norway['Perceptions of corruption']
}
# Plotting a pie chart of the factors' values
# - factors.values(): the sizes of each pie slice
# - labels=factors.keys(): assigns labels to each slice
# - autopct='%1.1f%%': displays the percentage value on each slice with 1 decimal
# - startangle=140: rotates the start of the pie chart for better layout
plt.pie(factors.values(), labels=factors.keys(), autopct='%1.1f%%', startangle=140)
# Adding a title to the chart
plt.title('Factor Contributions to Happiness Score (Norway)')
# Setting equal aspect ratio so that the pie is drawn as a circle
plt.axis('equal')
# Displaying the plot
plt.show()
```

```python
# Extract the row of data corresponding to Norway using filtering
# .iloc[0] is used to get the first (and only) row as a Series object
Norway = data[data['Country or region'] == 'Norway'].iloc[0]
# Creating a dictionary of key happiness factors for Norway with their corresponding values
# These factors are used to analyze the components contributing to Norway's happiness score
factors = {
    'GDP': Norway['GDP per capita'],
    'Social Support': Norway['Social support'],
    'Life Expectancy': Norway['Healthy life expectancy'],
    'Freedom': Norway['Freedom to make life choices'],
    'Generosity':Norway['Generosity'],
    'Corruption': Norway['Perceptions of corruption']
}

# Setting the "explode" values to separate each slice slightly from the center
# 0.1 is used to pull out each slice equally for visual emphasis
explode = [0.1] * len(factors) 

# Plotting a pie chart using the factor values and labels
# - labels=factors.keys() sets the labels as the dictionary keys
# - autopct='%1.1f%%' shows percentages rounded to 1 decimal place
# - startangle=140 rotates the start of the first slice for better layout
# - explode adds space between slices
# - shadow=True gives a 3D shadow effect
plt.pie(factors.values(), labels=factors.keys(), autopct='%1.1f%%',
        startangle=140, explode=explode, shadow=True)

# Setting the title of the pie chart
plt.title('Exploded Pie Chart: Factor Contributions to Happiness Score (Norway)')

# Making sure the pie chart is a perfect circle (not oval)
plt.axis('equal')
# Displaying the pie chart
plt.show()
```

```python
# BAR GRAPH

plt.figure(figsize=(8,5))

# Select the data row for Finland from the dataset
finland = data[data['Country or region'] == 'Finland'].iloc[0]

# Create a dictionary of key factors contributing to Finland's happiness score
factors = {
    'GDP': finland['GDP per capita'],                          
    'Social Support': finland['Social support'],               
    'Life Expectancy': finland['Healthy life expectancy'],     
    'Freedom': finland['Freedom to make life choices'],       
    'Generosity': finland['Generosity'],                        
    'Corruption': finland['Perceptions of corruption']         
}

# Prepare keys and values for plotting
factor_names = list(factors.keys())
factor_values = list(factors.values())

# Plot a bar graph to visualize the contributions of these factors
plt.bar(factor_names, factor_values, color='#aaccdd')

# Add a title to the plot
plt.title('Factor Contributions to Happiness Score (Finland)')

# Label the x-axis and y-axis
plt.xlabel('Factors')
plt.ylabel('Score Contribution')

# Display the values on top of each bar
for i, value in enumerate(factor_values):
    plt.text(i, value + 0.01, f'{value:.2f}', ha='center', va='bottom')

# Show the bar chart
plt.show()
```
